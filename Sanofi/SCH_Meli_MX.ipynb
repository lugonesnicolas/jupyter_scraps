```python
# Importacion de librerias necesarias
import BeOnFastCode as fc
import requests
import re
from bs4 import BeautifulSoup

# Proxy y datos como header para la coneccion

url = "https://listado.mercadolibre.com.mx/cansancio^#D^[A:cansancio^]"

payload = {}
headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:131.0) Gecko/20100101 Firefox/131.0',
  'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/png,image/svg+xml,*/*;q=0.8',
  'Accept-Language': 'es-AR,es;q=0.8,en-US;q=0.5,en;q=0.3',
  'Accept-Encoding': 'gzip, deflate, br, zstd',
  'DNT': '1',
  'Sec-GPC': '1',
  'Alt-Used': 'listado.mercadolibre.com.mx',
  'Connection': 'keep-alive',
  'Cookie': '_d2id=5f289d0b-ad48-43a1-b2f2-15b021462d41; _gcl_au=1.1.1134466840.1727046311; _ga_HEZRL7JLMR=GS1.1.1727046311.1.0.1727046315.56.0.0; _ga=GA1.1.35196849.1727046312; _csrf=dJ2sRK0kGsSL0MSD6Yzj7bs2; main_domain=; main_attributes=; categories=; last_query=cansancio; category=; backend_dejavu_info=j^%^3A^%^7B^%^7D; c_ui-navigation=6.6.87; backend_dejavu_info=j%3A%7B%7D; categories=; category=; last_query=cansancio%5E; main_attributes=; main_domain=',
  'Upgrade-Insecure-Requests': '1',
  'Sec-Fetch-Dest': 'document',
  'Sec-Fetch-Mode': 'navigate',
  'Sec-Fetch-Site': 'cross-site',
  'Sec-Fetch-User': '?1',
  'Priority': 'u=0, i'
}

# Consulta
response = requests.request("GET", url, headers=headers, data=payload)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")

products_list = soup.select("div.ui-search-result__wrapper") if soup.select("div.ui-search-result__wrapper") != None else ''

for product in products_list:
    title = product.select_one("a").text if product.select_one("a") != None else ''
    final_price = product.select_one("span.andes-money-amount--cents-superscript").text if product.select_one("span.andes-money-amount--cents-superscript") != None else '0'
    final_price = fc.just_price(price_str=final_price, dotted_price=True)
    old_price = product.select_one("s.andes-money-amount.andes-money-amount--previous.andes-money-amount--cents-dot").text if product.select_one("s.andes-money-amount.andes-money-amount--previous.andes-money-amount--cents-dot") != None else '0'
    old_price = fc.just_price(price_str=old_price, dotted_price=True)
    best_seller = product.select_one("span.poly-component__highlight").text if product.select_one("span.poly-component__highlight") != None else ''

    
    
    print(f"Title: {title}")
    print(f"FinalPrice: {final_price}")
    print(f"OldPrice: {old_price}")
    print(f"BestSeller: {best_seller}\n")

# # Parseo
# count = 0
# products_list = soup.select("test")
# for product in products_list:
#     if count < 10:
#         # Producto
```
