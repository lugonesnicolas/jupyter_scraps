```python
import requests
import json
import time
import re
from bs4 import BeautifulSoup
import time
```

```python
url = "https://www.masonline.com.ar/leche-la-serenisima-larga-vida-1-l/p"
```

```python
response = requests.get(url)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    print("La solicitud se realizó correctamente")
else:
    print("Hubo un error al realizar la solicitud")
```

```python
print(soup)
```

```python
title = soup.find("span", class_="vtex-store-components-3-x-productBrand")
q_images = soup.find_all(class_="absolute absolute--fill b--solid b--muted-2 bw0 vtex-store-components-3-x-carouselThumbBorder vtex-store-components-3-x-carouselThumbBorder--cl-imgProduct")
description = soup.find_all(".vtex-store-components-3-x-productDescriptionText > div:nth-child(1)")
```

```python
final_price = soup.find("div", class_="valtech-gdn-dynamic-product-0-x-dynamicProductPrice")
old_price = soup.find("span", class_='mt4.valtech-gdn-dynamic-product-0-x-weighableListPrice')
if final_price == None:
    final_price = soup.find("div", class_='mt4 valtech-gdn-dynamic-product-0-x-weighableListPrice')
final_price = 0 if final_price == None else final_price
old_price = 0 if old_price == None else old_price
```

```python
print(f"Title:{title.text}")
print(f"QImages:{len(q_images)-1}")
print(f"FinalPrice:{final_price.text}")
print(f"OldPrice:{old_price}")
print(f"Description:{description}")
```

```python
start_time = time.time()
for i in range (0, 301):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        print("La solicitud se realizó correctamente")
        title = soup.find("span", class_="vtex-store-components-3-x-productBrand")
        q_images = soup.find_all(class_="absolute absolute--fill b--solid b--muted-2 bw0 vtex-store-components-3-x-carouselThumbBorder vtex-store-components-3-x-carouselThumbBorder--cl-imgProduct")
        final_price = soup.find("div", class_="valtech-gdn-dynamic-product-0-x-dynamicProductPrice")
        old_price = soup.find("span", class_='mt4.valtech-gdn-dynamic-product-0-x-weighableListPrice')
        if final_price == None:
            final_price = soup.find("div", class_='mt4 valtech-gdn-dynamic-product-0-x-weighableListPrice')
        final_price = 0 if final_price == None else final_price
        old_price = 0 if old_price == None else old_price
        print(f"Title:{title.text}")
        print(f"QImages:{len(q_images)-1}")
        print(f"FinalPrice:{final_price.text}")
        print(f"OldPrice:{old_price}")
        print(f"\n")
    else:
        print("Hubo un error al realizar la solicitud")
end_time = time.time()
execution_time = end_time - start_time
print(f"Tiempo de ejecución: {execution_time} segundos")
```
