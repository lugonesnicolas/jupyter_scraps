```python
import requests
import json
import time
import re
```

```python
start_time = time.time()
```

```python
urls = ['https://www.masonline.com.ar/_v/segment/routing/vtex.store@2.x/product/35699/pure-de-tomate-great-value-520-gr/p?__pickRuntime=appsEtag%2Cblocks%2CblocksTree%2Ccomponents%2CcontentMap%2Cextensions%2Cmessages%2Cpage%2Cpages%2Cquery%2CqueryData%2Croute%2CruntimeMeta%2Csettings&__device=phone','https://www.masonline.com.ar/_v/segment/routing/vtex.store@2.x/product/173153/leche-uht-deslactosada-check-1-lt/p?__pickRuntime=appsEtag%2Cblocks%2CblocksTree%2Ccomponents%2CcontentMap%2Cextensions%2Cmessages%2Cpage%2Cpages%2Cquery%2CqueryData%2Croute%2CruntimeMeta%2Csettings&__device=tablet',
        'https://www.masonline.com.ar/_v/segment/routing/vtex.store@2.x/product/181562/leche-la-serenisima-larga-vida-1-l/p?__pickRuntime=appsEtag%2Cblocks%2CblocksTree%2Ccomponents%2CcontentMap%2Cextensions%2Cmessages%2Cpage%2Cpages%2Cquery%2CqueryData%2Croute%2CruntimeMeta%2Csettings&__device=tablet']
```

```python
def get_data(url:str):
    response = requests.get(url)
    data = response.json()
    product = data["queryData"]
    product_data = product[0]["data"]
    product_data_json = json.loads(product_data)
    # print(product_data_json)
    # print(type(product_data_json))
    title = product_data_json['product']['productName']
    link = product_data_json['product']['link']
    link = link.replace("https://portal.vtexcommercestable.com.br/", "")
    url = f"https://www.masonline.com.ar/{link}" 
    description = product_data_json['product']['description']
    brand = product_data_json['product']['brand']
    sku = product_data_json['product']['productId']
    ean = product_data_json['product']['items'][0]['ean']
    final_price = product_data_json['product']['priceRange']['sellingPrice']['highPrice']
    old_price = product_data_json['product']['priceRange']['listPrice']['highPrice']
    images_data = product_data_json['product']['items'][0]['images']
    print(f"Title:{title}")
    print(f"Description:{description}")
    print(f"Link:{url}")
    print(f"Brand:{brand}")
    print(f"SKU:{sku}")
    print(f"EAN:{ean}")
    print(f"Final Price:{final_price}")
    print(f"Old Price:{old_price}")
    print(f"Imagenes secundarias:{len(images_data)-1}")
    print(f"\n")
    return product_data_json
```

```python
for url in urls:
    product_data_json = get_data(url)
```

```python
end_time = time.time()
execution_time = end_time - start_time
print(f"Tiempo de ejecución: {execution_time} segundos")
```
