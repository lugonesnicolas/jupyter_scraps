```python
import requests
```

```python
url = "https://api.retailrocket.net/api/2.0/recommendation/Search/602bba6097a5281b4cc438c9/?&phrase=antigripal&session=668573720b0938896f284178&pvid=217082080063480&isDebug=false&format=json"

payload = {}
headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:127.0) Gecko/20100101 Firefox/127.0',
  'Accept': '*/*',
  'Accept-Language': 'es-AR,es;q=0.8,en-US;q=0.5,en;q=0.3',
  'Accept-Encoding': 'gzip, deflate, br, zstd',
  'Origin': 'https://salcobrand.cl',
  'DNT': '1',
  'Sec-GPC': '1',
  'Connection': 'keep-alive',
  'Referer': 'https://salcobrand.cl/',
  'Sec-Fetch-Dest': 'empty',
  'Sec-Fetch-Mode': 'cors',
  'Sec-Fetch-Site': 'cross-site',
  'TE': 'trailers'
}

response = requests.request("GET", url, headers=headers, data=payload)
data_grid = response.json()
```

```python
print(data_grid[0])
```

```python
for i in data_grid:
    title = i.get("Name", "") if i.get("Name", "") != None else ""
    brand = i.get("Vendor", "") if i.get("Vendor", "") != None else ""
    final_price = float(i.get("Price", "")) if i.get("Price", "") != None else 0
    old_price = float(i.get("OldPrice", "")) if i.get("OldPrice", "") != None else 0
    
    print(f"Title: {title}")
    print(f"Brand: {brand}")
    print(f"FinalPrice: {final_price}")
    print(f"OldPrice: {old_price}\n")
```
