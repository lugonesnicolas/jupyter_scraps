```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from time import sleep 
import json
```

```python
url = "https://salcobrand.cl/search_result?query=ibuprofeno"
```

```python
driver = webdriver.Chrome()
driver.get(url)
response = driver.execute_script("return document.status")
html = driver.page_source
driver.quit()
soup = BeautifulSoup(html, "lxml")
print(f"Response: {response}")
```

```python
product_list = soup.find_all("div", class_="inner-product-box") if soup.find_all("div", class_="inner-product-box") != [] else None
```

```python
position = 0
for i in product_list:
    position += 1
    if position < 10:
        title = i.select_one("span.product-info.truncate").text if i.select_one("span.product-info.truncate") != None else ''
        brand = i.select_one("span.product-name.truncate").text if i.select_one("span.product-name.truncate") != None else ''
        final_price = i.select_one("div.sale-price span").text if i.select_one("div.sale-price span") != None else 0
        final_price = float(final_price.replace("$", "").strip()) if final_price != 0 else final_price
        old_price = i.select_one("span.price.selling").text if i.select_one("span.price.selling") != None else 0
        old_price = float(old_price.replace("$", "").replace("Precio farmacia:", "").strip()) if old_price != 0 else old_price

        print(f"Position: {position}")
        print(f"Title: {title}")
        print(f"Brand: {brand}")
        print(f"FinalPrice: {final_price}")
        print(f"OldPrice: {old_price}\n")
```
