```python
products_list = soup.select("div.sg-col-4-of-24.sg-col-4-of-12.s-result-item.s-asin.sg-col-4-of-16.sg-col.s-widget-spacing-small.sg-col-4-of-20")
```

```python
for i in range(0,100):
    for product in products_list:
        title = product.select_one("span.a-size-base-plus.a-color-base.a-text-normal").text if product.select_one("span.a-size-base-plus.a-color-base.a-text-normal") != None else ''
        final_price = product.select_one("span.a-price").span.text if product.select_one("span.a-price") != None else 0
        old_price = product.select_one("div.a-section.aok-inline-block span.a-price.a-text-price").span.text if product.select_one("div.a-section.aok-inline-block span.a-price.a-text-price") != None else 0
        brand = product.select_one("h2.a-size-mini.s-line-clamp-1").text if product.select_one("h2.a-size-mini.s-line-clamp-1") != None else ''
        asin = product["data-asin"]
        product_url = product.select_one("a.a-link-normal.s-underline-text.s-underline-link-text.s-link-style.a-text-normal")["href"] if product.select_one("a.a-link-normal.s-underline-text.s-underline-link-text.s-link-style.a-text-normal") != None else ''
        product_url = f"https://www.amazon.ae{product_url}" if product_url != '' else ''
        print(f"Title:{title}")
        print(f"FinalPrice:{final_price}")
        print(f"OldPrice:{old_price}")
        print(f"Asin:{asin}")
        print(f"Brand:{brand}")
        print(f"ProductUrl:{product_url}\n")
```
